## Fibonacci

### function

````
def fibonacci_n(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        a = 0
        b = 1
        for i in range(2, n + 1):
            c = a + b
            a = b
            b = c
        return b

print(fibonacci_n(9))
````

### procedure

````
def printFibonacci(n):
    a = 0
    b = 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b
    print()

printFibonacci(9)
````

## 2.9. Indexing and Slicing Array

Indexing dan slicing adalah teknik pada NumPy untuk mengambil data dari array. Indexing digunakan untuk mengambil satu elemen berdasarkan indeks, sedangkan slicing digunakan untuk mengambil beberapa elemen menggunakan rentang indeks. Teknik ini membantu mempermudah manipulasi dan analisis data.

````
# Simple Indexing

x = np.array([10, 20, 30, 40, 50])

print(x[2])
````

Pengindeksan sederhana digunakan untuk mengakses satu elemen dari array berdasarkan indeksnya. Dalam contoh ini, `x[2]` mengakses elemen ke-3 dari array `x`, yaitu nilai 30.

````
# Slicing

x = np.array([10, 20, 30, 40, 50])

# Slicing to get elements from index 1 to 3
subset = x[1:4]

print(subset)
````

Slicing digunakan untuk mengambil subset dari array dengan rentang indeks. Dalam contoh ini, `x[1:4]` mengakses
elemen mulai dari indeks 1 hingga 3 (indeks akhir tidak termasuk), menghasilkan array `[20, 30, 40]`.

````
# Boolean Indexing

x = np.array([10, 20, 30, 40, 50])

# Boolean mask to select elements greater than 25
mask = x > 25
print("Boolean result of masking:", mask)

subset = x[mask]
print("Result elemen:", subset)
````

Pengindeksan Boolean menggunakan boolean array (mask) untuk memilih elemen yang memenuhi kondisi tertentu.
Di sini, `x > 25` menghasilkan mask `[False, False, True, True, True]`,
yang kemudian digunakan untuk memilih elemen yang lebih besar dari 25.

````
# Fancy Indexing

x = np.array([10, 20, 30, 40, 50])

# Array of index
index = [1, 3]

subset = x[index]
print(subset)
````

Fancy Indeks digunakan untuk mengakses elemen array menggunakan indeks array. Dalam contoh ini, `x[index]`
memilih elemen pada indeks 1 dan 3 dari array `x`, yaitu 20 dan 40.

`x[1, ...]` mengakses semua elemen di baris kedua array 2D x, menghasilkan array `[4, 5, 6]`.

## 2.10. Copying (Copy) Array

Copying array pada NumPy digunakan untuk menyalin data dari sebuah array. Ada dua jenis yaitu shallow copy, yang masih berbagi data dengan array asli, dan deep copy, yang membuat salinan terpisah sehingga perubahan pada salinan tidak mempengaruhi array asli.

````
# Syntax`view()`

x = np.array([1, 2, 3, 4, 5])

# Create a shallow copy using view()
y = x.view()

# Changes the first element of the shallow copy
y[0] = 99

print("Original array:", x)
print("Shallow copy:", y)
````

Dalam contoh di atas, array `and` adalah salinan dangkal dari array `x`, yang dibuat menggunakan `view()`. Ketika
elemen pertama dalam array `and` diubah, perubahan tersebut juga tercermin dalam array `x`, karena
keduanya berbagi data yang sama.

````
# Syntax`copy()`

x = np.array([1, 2, 3, 4, 5])

# Create a deep copy using copy()
y = x.copy()

# Changes the first element of the deep copy
y[0] = 99

print("Original array:", x)
print("Deep copy:", y)
````

Di sini, array `and` adalah salinan mendalam dari array `x`, yang dibuat menggunakan `copy()`. Ketika elemen pertama
dalam array `and` diubah, array asli `x` tidak terpengaruh, karena salinan mendalam membuat
salinan independen yang tidak berbagi data dengan array asli.

### 3.2. Creating an Empty Graph on Networkx

Pada NetworkX terdapat dua jenis graf dasar yang sering digunakan, yaitu Graph dan DiGraph. Graph digunakan untuk membuat graf tidak berarah (undirected graph), yaitu hubungan antar node tidak memiliki arah sehingga kedua node saling terhubung secara simetris.

Sedangkan DiGraph digunakan untuk membuat graf berarah (directed graph), yaitu hubungan antar node memiliki arah tertentu sehingga hubungan dari node A ke node B tidak selalu berarti ada hubungan dari node B ke node A.

````
# Import the library used
import matplotlib.pyplot as plt
import networkx as nx

# Node positions used in this module
pos = {'A': [ 0.93888004, -0.0449161 ],
       'B': [-0.73226196, -0.53345128],
       'C': [0.29334637, 1. ],
       'D': [-0.79991428, 0.56134392],
       'E': [ 0.29994983, -0.98297654]}

# Support function for printing graphs
def show_graph(G, pos=None, title='') :
  if pos == None :
    pos = nx.spring_layout(G)

    nx.draw(
        G,
        pos,
        with_labels=True,
        node_color='red',
        node_size=2000,
        font_color="white",
        font_weight="bold",
        width=5
    )
    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(
        G,
        pos,
        edge_labels=edge_labels,
        font_color='blue',
        font_weight="bold",
        font_size=12,
    )

    plt.margins(0.2)
    plt.title(title)
    plt.show()
````

#### Definition `Graph` empty

````
G_undirected = nx.Graph()

# Print the graph
show_graph(G_undirected,
title="Undirected Graph Kosong")
````

Hubungan antar node tidak punya arah.

Kalau ada edge antara A dan B, berarti A terhubung ke B dan B juga ke A.

Hubungannya dua arah/simetris.

#### Definition ``DiGraph'' empty

````
G_directed = nx.DiGraph()

# Print the graph
show_graph(G_directed,
title="Directed Graph Kosong")
````

Hubungan antar node punya arah.

Kalau ada edge dari A → B, belum tentu ada edge dari B → A.

Hubungannya satu arah.

nx.Graph() digunakan untuk membuat graf tidak berarah (undirected graph) yang dimana hubungan antar node bersifat dua arah. Sedangkan, nx.DiGraph() digunakan untuk membuat graf berarah (directed graph) dimana hubungan antar node memiliki arah tertentu.

### 3.3. Add Node On the Graph




#### Add node on undirected graph

````
import networkx as nx
import matplotlib.pyplot as plt

# Buat undirected graph
G_undirected = nx.Graph()

# Tambah node satu per satu
G_undirected.add_node('A')
G_undirected.add_node('B')
G_undirected.add_node('C')
G_undirected.add_node('D')
G_undirected.add_node('E')

# Atau pakai add_nodes_from
node_data = ['A', 'B', 'C', 'D', 'E']
G_undirected.add_nodes_from(node_data)

# Tentukan posisi node (spring layout)
pos = nx.spring_layout(G_undirected)

# Tampilkan graf
nx.draw(G_undirected, pos, with_labels=True, node_color='lightgreen', node_size=1500)
plt.title("Add Node Undirected Graph")
plt.show()

# Jumlah node
print("Jumlah node:", G_undirected.number_of_nodes())
````

#### Add node on directed graph

````
import networkx as nx
import matplotlib.pyplot as plt

# Buat directed graph
G_directed = nx.DiGraph()

# Tambah node satu per satu
G_directed.add_node('A')
G_directed.add_node('B')
G_directed.add_node('C')
G_directed.add_node('D')
G_directed.add_node('E')

# Atau pakai add_nodes_from
node_data = ['A', 'B', 'C', 'D', 'E']
G_directed.add_nodes_from(node_data)

# Tentukan posisi node (spring layout)
pos = nx.spring_layout(G_directed)

# Tampilkan graf
nx.draw(G_directed, pos, with_labels=True, node_color='skyblue', node_size=1500, arrowstyle='-|>')
plt.title("Add Node Directed Graph")
plt.show()

# Jumlah node
print("Jumlah node:", G_directed.number_of_nodes())
````

Kedua kode itu berfungsi untuk membuat graf dengan node A, B, C, D, dan E.

````
print("Tetangga G_undirected")
for node in G_undirected.nodes():
  # Neighbors of a given node
  adj = list(G_undirected.adj[node])
  print(f"Tetangga {node}: {adj}")
````

#### `.adj()`on directed graph

````
print("Tetangga G_directed")
for node in G_directed.nodes():
  # Neighbors of a given node
  adj = list(G_directed.adj[node])
  print(f"Tetangga {node}: {adj}")
````

#### `.degree()`

#### `.adj()` on undirected graph

````
print("G_undirected Neighbor")
for node in G_undirected.nodes():
    neighbors = list(G_undirected.neighbors(node))
    print(f"Neighbor {node}: {neighbors}")
````

````
print("G_directed neighbor")
for node in G_directed.nodes():
    neighbors = list(G_directed.neighbors(node))
    print(f"Neighbor {node}: {neighbors}")
````

#### `.in_degree()`

````
import networkx as nx
import matplotlib.pyplot as plt

# Buat graf undirected
G_undirected = nx.Graph()

# Tambah node
nodes = ['A', 'B', 'C', 'D', 'E']
G_undirected.add_nodes_from(nodes)

# Tambah weighted edge
edge_weight_data = [('A', 'B', 10), ('B', 'C', 7), ('C', 'D', 4), ('D', 'B', 8), ('E', 'D', 3)]
G_undirected.add_weighted_edges_from(edge_weight_data)

# Tentukan posisi node
pos = nx.spring_layout(G_undirected)

# Gambar graf
nx.draw(G_undirected, pos, with_labels=True, node_color='lightgreen', node_size=1500)

# Tambahkan label weight
edge_labels = nx.get_edge_attributes(G_undirected, 'weight')
nx.draw_networkx_edge_labels(G_undirected, pos, edge_labels=edge_labels)

plt.title("Add Weight Undirected Graph")
plt.show()
````

#### Add weight on directed graph

````
import networkx as nx
import matplotlib.pyplot as plt

# Buat directed graph
G_directed = nx.DiGraph()

# Tambah node
nodes = ['A', 'B', 'C', 'D', 'E']
G_directed.add_nodes_from(nodes)

# Tambah weighted edges (cukup salah satu cara)
edge_weight_data = [('A', 'B', 10), ('B', 'C', 7), ('C', 'D', 4), ('D', 'B', 8), ('E', 'D', 3)]
G_directed.add_weighted_edges_from(edge_weight_data)

# Tentukan posisi node
pos = nx.spring_layout(G_directed)

# Gambar graf
nx.draw(G_directed, pos, with_labels=True, node_color='skyblue', node_size=1500,
        arrowstyle='-|>', arrowsize=20)

# Tambahkan label weight
edge_labels = nx.get_edge_attributes(G_directed, 'weight')
nx.draw_networkx_edge_labels(G_directed, pos, edge_labels=edge_labels)

plt.title("Add Weight Directed Graph")
plt.show()
````

### 3.8. Graph Analysis

#### `.connected_components()`
Fungsi ini digunakan untuk menemukan semua komponen terhubung dalam graf tak berarah, yang berarti setiap simpul dalam suatu kelompok dapat dijangkau oleh simpul lain dalam kelompok yang sama.

````
import networkx as nx
import matplotlib.pyplot as plt

# Buat graf undirected
G = nx.Graph()

# Tambah edges (node otomatis ikut dibuat)
G.add_edges_from([('B', 'D'), ('B', 'E'), ('C', 'A')])

# Ambil connected components
components = list(nx.connected_components(G))

# Tampilkan components sebagai list of sets
print(components)  # Output: [{'D', 'E', 'B'}, {'A', 'C'}]

# Tentukan posisi node untuk visualisasi
pos = nx.spring_layout(G)

# Gambar graf
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=1500)
plt.title("Graph Connected Component")
plt.show()
````

#### `.clustering()`
Fungsi ini mengukur koefisien pengelompokan dari setiap node dalam grafik. Koefisien pengelompokan adalah ukuran lokal yang menggambarkan seberapa dekat tetangga antar node satu sama lain terhubung.

````
import networkx as nx
import matplotlib.pyplot as plt

# Buat graf undirected
G = nx.Graph()

# Tambah edges (node otomatis ikut dibuat)
edge_data = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'B'), ('E', 'D')]
G.add_edges_from(edge_data)

# Hitung clustering coefficient tiap node
clustering_coefficients = nx.clustering(G)
print(clustering_coefficients)  # Output: dictionary node: nilai clustering

# Tentukan posisi node untuk visualisasi
pos = nx.spring_layout(G)

# Gambar graf
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=1500)
plt.title("Graph Clustering")
plt.show()
````

#### `.all_shortest_path()`
Fungsi ini digunakan untuk menemukan semua jalur terpendek antara dua node dalam graf. Fungsi ini mencari semua rute yang memiliki nilai tepi terkecil dari kedua node tertentu.

````
import networkx as nx
import matplotlib.pyplot as plt

# Buat graf undirected
G = nx.Graph()

# Tambah weighted edges
edge_weight_data = [('A', 'B', 10), ('B', 'C', 7), ('C', 'D', 4), ('D', 'B', 8), ('E', 'D', 3)]
G.add_weighted_edges_from(edge_weight_data)

# Hitung semua shortest path untuk semua node
# Jika ingin mempertimbangkan bobot, ganti dengan nx.all_pairs_dijkstra_path(G, weight='weight')
shortest_paths = dict(nx.all_pairs_shortest_path(G))

# Print shortest paths
print(shortest_paths)

# Tentukan posisi node
pos = nx.spring_layout(G)

# Gambar graf
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1500)

# Tambahkan label bobot edge
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title("Graph All Shortest Path")
plt.show()
````

##### `spring_layout()`
Fungsi ini menggunakan model spring, di mana simpul-simpul dihubungkan oleh spring. Simpul-simpul akan ditempatkan cenderung lebih dekat satu sama lain, sedangkan simpul-simpul yang tidak terhubung akan ditempatkan lebih jauh.

````
import networkx as nx
import matplotlib.pyplot as plt

# Buat graf
G = nx.Graph()

# Tambah edges (node otomatis dibuat)
edges = [('A', 'B'), ('A', 'C'), ('A', 'D'),
         ('B', 'C'), ('B', 'D'), ('C', 'D')]
G.add_edges_from(edges)

# Atur posisi node dengan spring_layout
pos = nx.spring_layout(G)  # Node akan terdistribusi rapi otomatis

# Gambar graf
nx.draw(G, pos, with_labels=True, node_size=1500, node_color='red', edge_color='black', width=2)

plt.title("Graph Spring")
plt.show()
````

#### `shell_layout()`
Fungsi ini menempatkan node dalam beberapa shell. Setiap shell adalah lingkaran yang mengelilingi pusat, dan node tertentu dikelompokkan dalam cangkang yang berbeda. Parameter `nlist` mendefinisikan kelompok node yang akan ditempatkan di lapisan tertentu.

````
import networkx as nx
import matplotlib.pyplot as plt

# Buat graf contoh
G_shell = nx.Graph()

# Tambah edges (node otomatis dibuat)
edges = [
    ('A', 'B'), ('A', 'C'), ('A', 'D'),
    ('B', 'E'), ('B', 'F'), ('C', 'G'),
    ('D', 'H'), ('D', 'I'), ('E', 'J'),
    ('F', 'K'), ('G', 'L'), ('H', 'M')
]
G_shell.add_edges_from(edges)

# Tentukan posisi node menggunakan shell_layout
pos = nx.shell_layout(
    G_shell,
    nlist=[
        ['A'],                 # Lapis 1
        ['B', 'C', 'D'],       # Lapis 2
        ['E', 'F', 'G'],       # Lapis 3
        ['H', 'I', 'J', 'K', 'L', 'M']  # Lapis 4
    ]
)

# Gambar graf
nx.draw(G_shell, pos, with_labels=True, node_size=1200, node_color='lightblue', edge_color='black', width=2)
plt.title("Graph Shell Layout")
plt.show()
````

#### `circular_layout()`
Fungsi ini menempatkan semua node dalam grafik pada sebuah lingkaran dengan jarak yang sama dari pusat. Posisi setiap node dalam lingkaran ditentukan secara merata sehingga membentuk susunan simetris.

````
import networkx as nx
import matplotlib.pyplot as plt

# Buat graph
G = nx.Graph()
G.add_edges_from([('C','D'), ('C','A'), ('C','B'), ('B','A'), ('B','C'), ('B','D'), ('D', 'A')])

# Atur posisi node di lingkaran
pos = nx.circular_layout(G)

# Gambar graph
nx.draw(G, pos, with_labels=True, node_color='red', node_size=800, font_color='white', font_weight='bold', edge_color='black', width=2)

plt.title("Graph Shell")
plt.show()
````

#### 3.9.3. Function show_graph For Graph Visualization

Fungsi `show_graph` adalah fungsi pendukung yang dirancang untuk memvisualisasikan grafik yang dibuat dengan NetworkX. Fungsi ini menyediakan cara yang mudah dan fleksibel untuk menampilkan grafik dengan berbagai opsi. Semua grafik yang divisualisasikan dalam modul praktikum ini ditampilkan menggunakan fungsi `show_graph`.

````
import matplotlib.pyplot as plt
import networkx as nx

# Support function for printing graphs
def show_graph(G, pos=None, title=''):
    # Create a pos if a pos is not given
    if pos is None:
        pos = nx.spring_layout(G)

    # Function to draw nodes
    nx.draw(
        G,                    # Graf NetworkX
        pos,                  # Node position
        with_labels=True,     # Displays the node name
        node_color='red',     # Node color
        node_size=2000,       # Node size
        font_color="white",   # Node label font color
        font_weight="bold",   # Node label font thickness
        width=5               # Edge line thickness
    )

    # Takes edge labels if there is weight
    edge_labels = nx.get_edge_attributes(G, 'weight')

    # Function to draw edge labels
    nx.draw_networkx_edge_labels(
        G,
        pos,
        edge_labels=edge_labels, # Data weight
        font_color='blue',       # Edge label font color
        font_weight="bold",      # Edge label font weight
        font_size=12             # Edge label font size
    )

    plt.margins(0.2)  # Provides margin to the plot
    plt.title(title)   # Displays the graph title if given
    plt.show()         # Display graphs using matplotlib
````

#### 5.1.2. Implementation of BFS On Directed Graph

BFS pada graf berarah dimulai dari simpul awal (simpul sumber) dan menjelajahi semua tetangga yang dapat dijangkau langsung dari simpul tersebut. Setelah semua tetangga pada tingkat pertama telah dijelajahi, BFS melanjutkan ke tetangga dari semua simpul pada tingkat ini, tingkat berikutnya, dan seterusnya. Pada graf berarah, BFS hanya mengikuti arah tepi saat menjelajahi graf. BFS pada graf berarah dapat digunakan untuk menentukan jalur terpendek dari satu simpul ke simpul lainnya; semua simpul lain dalam graf tidak diberi bobot.

##### Function to print plots

````
import networkx as nx
import matplotlib.pyplot as plt

# Support function for printing graphs
def show_graph(G, pos=None, title='', labels=None):

    if pos is None:
        pos = nx.spring_layout(G)

    if labels is None:
        labels = {node: node for node in G.nodes()}

    # Gambar node
    nx.draw(
        G,
        pos,
        with_labels=True,
        labels=labels,
        node_color='red',
        node_size=2000,
        font_color="white",
        font_weight="bold",
        width=2
    )

    edge_labels = nx.get_edge_attributes(G, 'weight')

    # Gambar weight
    nx.draw_networkx_edge_labels(
        G,
        pos,
        edge_labels=edge_labels,
        font_color='blue',
        font_weight="bold",
        font_size=12
    )

    plt.margins(0.2)
    plt.title(title)
    plt.show()
````

#8.1. Implementation Hill Climbing

Hill Climbing merupakan metode pencarian berbasis heuristik yang bertujuan memperoleh solusi optimal dari suatu ruang pencarian. Algoritma ini bekerja dengan mengeksplorasi titik-titik yang ada untuk menemukan nilai paling maksimal atau minimal, sesuai dengan kebutuhan permasalahan yang dihadapi.

## Implementation Example Hill Climbing
[Traveling Salesman Problem] One person salesman must visit 8 cities, where each city can only be visited once. The goal is to find a route with the minimum total distance from the distribution of cities in Figure 16.

image.png

### Import library which is used

````
# Import library
import random
import math

# Calculate the distance between 2 points using the Euclidean formula
def distance(city1, city2):
    return math.sqrt((city1[0] - city2[0]) ** 2 + (city1[1] - city2[1]) ** 2)

# Add up all the distances between points on the tour route
def total_distance(tour, cities):
    dist = 0
    for i in range(len(tour)):
        dist += distance(cities[tour[i]], cities[tour[(i + 1) % len(tour)]])
    return dist

# Provides all possible route combinations by swapping two points
def get_neighbors(tour):
    neighbors = []
    for i in range(len(tour)):
        for j in range(i + 1, len(tour)):
            neighbor = tour[:]
            neighbor[i], neighbor[j] = neighbor[j], neighbor[i]
            neighbors.append(neighbor)
    return neighbors


# =========================
# Hill Climbing Algorithm
# =========================

# Declaration of all city points
cities = [
    (0, 0), (1, 3), (4, 3), (6, 1),
    (3, 0), (2, 2), (5, 5), (7, 7)
]

# initial solution
current_solution = list(range(len(cities)))
random.shuffle(current_solution)

# initial distance
current_distance = total_distance(current_solution, cities)

# loop until no better solution
while True:
    neighbors = get_neighbors(current_solution)

    best_neighbor = min(neighbors, key=lambda tour: total_distance(tour, cities))
    best_distance = total_distance(best_neighbor, cities)

    if best_distance >= current_distance:
        break

    current_solution = best_neighbor
    current_distance = best_distance

# output
print("Best solution (city order):", current_solution)
print("Total best distance:", current_distance)
````

Algoritma Hill Climbing tidak selalu mampu menghasilkan solusi yang optimal karena cenderung terjebak pada optimum lokal, yaitu solusi terbaik di area sekitarnya, tetapi belum tentu merupakan solusi terbaik secara keseluruhan (optimum global). Oleh sebab itu, pada permasalahan yang kompleks, algoritma ini sering kali gagal menemukan solusi paling optimal tanpa bantuan metode tambahan seperti simulated annealing atau genetic algorithm.

Simulated Annealing merupakan algoritma optimasi yang mengadopsi prinsip termodinamika, khususnya proses pemanasan dan pendinginan material.

Berbeda dengan metode Hill Climbing yang cenderung kaku karena hanya menerima perbaikan solusi secara langsung, Simulated Annealing memiliki fleksibilitas untuk menerima solusi yang kurang optimal pada tahap awal. Mekanisme ini memungkinkan algoritma keluar dari jebakan local optimum guna mencapai hasil terbaik secara menyeluruh (global optimum).

## 9.1. How Algorithms Work Simulated Annealing

1. **Inisialisasi**: Buat solusi awal acak (urutan kota).
2. **Parameter suhu**: Tentukan suhu awal besar, *cooling rate*, dan batas berhenti.
3. **Perulangan**: Selama suhu > batas, suhu diturunkan tiap iterasi.
4. **Solusi baru**: Ubah solusi (misalnya tukar dua kota).
5. **Evaluasi**:

   * Jika lebih baik → diterima.
   * Jika lebih buruk → bisa diterima dengan probabilitas tertentu (Metropolis).
6. **Update terbaik**: Simpan solusi terbaik selama proses.
7. **Hasil akhir**: Berhenti saat suhu rendah, keluarkan solusi terbaik.

### **Simulated Annealing Algorith**

The algorithm will store 2 solution values. Best solution will save the best solution, and current solution will save the solution to do further exploration to avoid local  optimum.

````
import random
import math

def total_distance(tour, cities):
    distance = 0
    for i in range(len(tour)):
        x1, y1 = cities[tour[i]]
        x2, y2 = cities[tour[(i + 1) % len(tour)]]  # kembali ke awal
        distance += math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return distance

def random_swap(tour):
    new_tour = tour[:]
    i, j = random.sample(range(len(tour)), 2)
    new_tour[i], new_tour[j] = new_tour[j], new_tour[i]
    return new_tour

def simulated_annealing(cities, initial_temp, cooling_rate, stopping_temp):
    # initialize the initial solution
    current_solution = list(range(len(cities)))
    random.shuffle(current_solution)

    # hitung jarak awal
    current_distance = total_distance(current_solution, cities)

    # set best solution
    best_solution = current_solution[:]
    best_distance = current_distance

    # suhu awal
    temp = initial_temp

    while temp > stopping_temp:

        # solusi baru
        new_solution = random_swap(current_solution)
        new_distance = total_distance(new_solution, cities)

        # seleksi solusi
        if new_distance < current_distance:
            current_solution = new_solution[:]
            current_distance = new_distance
        else:
            if random.random() < math.exp((current_distance - new_distance) / temp):
                current_solution = new_solution[:]
                current_distance = new_distance

        # update best
        if current_distance < best_distance:
            best_solution = current_solution[:]
            best_distance = current_distance

        # turunkan suhu
        temp *= cooling_rate

    return best_solution, best_distance


# data kota
cities = [
    (0, 0), (1, 3), (4, 3), (6, 1),
    (3, 0), (2, 2), (5, 5), (7, 7)
]

# parameter
init = 10000
Cooling = 0.995
Stop = 1

# jalankan
best_solution, best_distance = simulated_annealing(cities, init, Cooling, Stop)

print("Best solution (city order):", best_solution)
print("Total best distance:", best_distance)
````

`if random.random() < math.exp((current_distance - new_distance)/temp)`
adalah inti Simulated Annealing.

* Jika solusi baru lebih buruk, masih bisa diterima dengan probabilitas (Metropolis Criterion).
* `random.random()` menghasilkan angka 0–1.
* `math.exp(...)` menentukan peluang penerimaan berdasarkan suhu (*temp*).

Tujuannya agar tidak terjebak di **local optimum**.
Berbeda dengan Hill Climbing, Simulated Annealing bisa menerima solusi lebih buruk di awal (suhu tinggi) untuk eksplorasi, lalu makin selektif saat suhu menurun, sehingga peluang mencapai **global optimum** lebih besar.

[Masalah Pedagang Keliling Berganda] Ada dua orang pedagang yang ingin mengunjungi 8 kota. Setiap pedagang hanya diperbolehkan mengunjungi setiap kota sekali, dan tidak ada kota yang dikunjungi lebih dari sekali oleh pedagang kedua. Koordinat dari 8 kota tersebut telah ditentukan berdasarkan titik-titik yang ada.

Buat solusi untuk menentukan rute perjalanan pedagang kedua menggunakan pendekatan algoritma Simulated Annealing. Rute yang dihasilkan harus meminimalkan total jarak yang ditempuh oleh kedua pedagang bersama-sama.

````
import random
import math

# hitung jarak euclidean
def distance(city1, city2):
    return math.sqrt((city1[0] - city2[0]) ** 2 + (city1[1] - city2[1]) ** 2)

# hitung jarak 1 rute
def route_distance(tour, cities):
    dist = 0
    for i in range(len(tour)):
        dist += distance(cities[tour[i]], cities[tour[(i + 1) % len(tour)]])
    return dist

# total jarak 2 salesman
def total_distance(solution, cities):
    tour, split = solution
    r1 = tour[:split]
    r2 = tour[split:]
    return route_distance(r1, cities) + route_distance(r2, cities)

# swap + ubah split
def random_swap(solution):
    tour, split = solution
    new_tour = tour[:]

    # swap dua kota
    i, j = random.sample(range(len(tour)), 2)
    new_tour[i], new_tour[j] = new_tour[j], new_tour[i]

    # ubah titik pembagi
    new_split = split + random.choice([-1, 1])
    new_split = max(1, min(len(tour) - 1, new_split))

    return new_tour, new_split

def simulated_annealing(cities, initial_temp, cooling_rate, stopping_temp):

    # solusi awal
    current_tour = list(range(len(cities)))
    random.shuffle(current_tour)

    split = len(cities) // 2
    current_solution = (current_tour, split)

    current_distance = total_distance(current_solution, cities)

    best_solution = current_solution
    best_distance = current_distance

    temp = initial_temp

    while temp > stopping_temp:

        # solusi baru
        new_solution = random_swap(current_solution)
        new_distance = total_distance(new_solution, cities)

        if new_distance < current_distance:
            current_solution = new_solution
            current_distance = new_distance
        else:
            if random.random() < math.exp((current_distance - new_distance) / temp):
                current_solution = new_solution
                current_distance = new_distance

        # update terbaik
        if current_distance < best_distance:
            best_solution = current_solution
            best_distance = current_distance
        temp *= cooling_rate

    return best_solution, best_distance


# DATA KOTA
cities = [
    (0, 0), (1, 3), (4, 3), (6, 1),
    (3, 0), (2, 2), (5, 5), (7, 7)
]

init = 10000
Cooling = 0.995
Stop = 1

best_solution, best_distance = simulated_annealing(cities, init, Cooling, Stop)

tour, split = best_solution

print("Salesman 1:", tour[:split])
print("Salesman 2:", tour[split:])
print("Total distance:", best_distance)
````

Solusi awal dimulai dari urutan kota yang diacak, lalu dibagi jadi dua rute pakai split. Setelah itu dihitung total jarak dari kedua rute. Untuk mencari solusi yang lebih baik, dilakukan pertukaran posisi kota dan perubahan split. Dengan Simulated Annealing, solusi yang lebih baik langsung diterima, sedangkan yang lebih buruk masih punya peluang diterima supaya tidak terjebak di local optimum, hingga akhirnya didapatkan rute dengan jarak minimum.
